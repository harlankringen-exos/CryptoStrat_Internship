```python
import pandas as pd
import numpy as np
```

```python
# open + close for the month of May
df = pd.read_csv('historic.csv')
df
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

fig = plt.figure()
ax = plt.axes()

plt.plot(df['time'], df['open'], color='b')
plt.plot(df['time'], df['close'], color='g')
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

fig = plt.figure()
ax = plt.axes()

plt.plot(df['time'], df['close'] - df['open'])
```
